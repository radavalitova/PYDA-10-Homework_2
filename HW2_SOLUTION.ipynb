## Задание 2

```python
# Високосные года делятся нацело на 4. Однако из этого правила есть исключение: столетия, которые не делятся нацело на 400, високосными не являются.
# В високосном годе 366 дней, в обычном 365.

# Если год не делится на 4, значит он обычный.
# Иначе надо проверить не делится ли год на 100.
# Если не делится, значит это не столетие и можно сделать вывод, что год високосный.
# Если делится на 100, значит это столетие и его следует проверить его делимость на 400.
# Если год делится на 400, то он високосный.
# Иначе год обычный.

year = int(input('Введите год: '))
x=year % 4
y=year % 100
z=year % 400

if x > 0:    
    print('Обычный год')
elif y > 0: 
    print('Високосный год')
elif z > 0: 
    print('Обычный год')
else:
    print ('Високосный год')
```

```python
# В задаче не уточнено для коробки №2, ВСЕ ли стороны должны быть меньше 50 см. 
#Потому заложила в условие, что коробка №2 имеет все 3 стороны менее 50см.

# Также если одна из сторон РАВНА 15 см и более (до 50см), 
#то это коробка №2 (иначе для коробки №1 нужно условие "менее или равно 15см")

# По условию задачи коробка для лыж, если ДЛИНА 200 см и более, но если покупатель ошибочно укажет 
#высоту 200см а остальные по 20 см, ему все равно понадобится коробка для лыж...
#Любая из сторон может быть 200см и более для лыжной коробки.


width = float(input('Введите ширину товара в сантиметрах: '))
length = float(input('Введите длину товара в сантиметрах: '))
height = float(input('Введите высоту товара в сантиметрах: '))

if (length >= 200) or (width >= 200) or (height >= 200):    
    print('Упаковка для лыж')
elif (width < 15) and (length < 15) and (height < 15): 
    print('Коробка №1')
elif (width < 50) and (length < 50) and (length < 50): 
    print('Коробка №2')
else:
    print ('Стандартная коробка №3')
```
